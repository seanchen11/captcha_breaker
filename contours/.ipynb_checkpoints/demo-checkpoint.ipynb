```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
def img_show(img):
    fig = plt.figure()
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
#     fig.savefig('test', bbox_inches='tight', pad_inches=0)
    plt.show()
```

```python
def img_save(img, i):
    fig = plt.figure()
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    path = 'digits/%s' % str(np.random.rand())[2:]
    fig.savefig(path, bbox_inches='tight', pad_inches=0)
    plt.show()
```

```python
img = cv.imread('0B7799.png')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# thresh = cv2.adaptiveThreshold(gray, 255, 1, 1, 11, 2)
ret, thresh = cv.threshold(gray, 127, 255, cv.THRESH_BINARY)
img2, contours, hierarchy = cv.findContours(thresh, cv.RETR_LIST, cv.CHAIN_APPROX_SIMPLE)
img_show(img)
for i, j in enumerate(contours):
    if (cv.contourArea(j) < 1000) and (cv.contourArea(j) > 250):
        [x,y,w,h] = cv.boundingRect(j)
        if  h > 10:
            cv.rectangle(img, (x - 10, y - 10), (x + w + 10, y + h + 10), (0, 255, 0), 2)
            roi = thresh[y-10:y+h+10, x-10:x+w+10]
            img_save(roi, i)
            roismall = cv.resize(roi, (10, 10))
            roismall = roismall.reshape((1, 100))
            roismall = np.float32(roismall)
img_show(img)
```
